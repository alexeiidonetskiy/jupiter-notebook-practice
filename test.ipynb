```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,4,6,8,10,12,14,16]

for i in range(2):
  x=list(map((lambda x: x **2), x))
  y=list(map((lambda i: i / -2), y))
  plt.plot(x,y)
  plt.show()
```
